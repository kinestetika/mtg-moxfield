```python
import random

def ad_naus(deck, life_lost_threshold=30):
    life_lost = 0
    fast_mana_total = 0
    total_cards = 0
    deck_copy = deck.copy()
    while (life_lost < life_lost_threshold):
        card_number = random.randint(0,len(deck_copy)-1)
        card = deck_copy[card_number]
        del deck_copy[card_number]
        if not card[0]:
            continue
        total_cards += 1
        life_lost += int(card[1])
        fast_mana_total += int(card[2])
    return [total_cards, fast_mana_total]

print(ad_naus(budget_deck, 30))
```
